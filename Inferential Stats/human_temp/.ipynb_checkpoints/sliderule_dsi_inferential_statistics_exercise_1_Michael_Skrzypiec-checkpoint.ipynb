## 1. Is the distribution of body temperatures normal?

```python
f, (ax1, ax2, ax3) = plt.subplots(3,1)
f.set_figheight(15)
f.set_figwidth(8)

sns.distplot(df.temperature, kde=False, ax=ax1, color='blue', label='Temp Counts')
sns.despine(ax=ax1)
ax1.set_xlim(95,102)
ax1.set_title('Body Temperature Sample Data Histogram')
ax1.set_ylabel('Counts')
ax1.set_xlabel('Body Temperature (F)')
ax1.legend()

sns.distplot(df.temperature, ax=ax2, kde=True, hist=False, color='blue', label='KDE')
sns.distplot(df.temperature, ax=ax2, kde=False, hist=False, fit=stats.norm, fit_kws={'color':'red'}, label='NORM FIT')
sns.despine(ax=ax2)
ax2.set_xlim(95,102)
ax2.set_title('Sample Data Normal Fit vs KDE')
ax2.set_ylabel('Frequency')
ax2.set_xlabel('Body Temperature (F)')
ax2.legend()


stats.probplot(df.temperature, dist="norm", plot=plt)
sns.despine(ax=ax3)
plt.title('Normal Q-Q Plot')


f.tight_layout()
plt.plot()
```

The three figures above examine the normality of the body temperature data.
### Figure 1 - Body Temperature Sample Data Histogram
A basic histogram of the data. The chart shows a gradual rise and fall between 96 and 101 degrees Fahrenheit with one peak between 98 and 99 degrees. This figure alone does not give strong enough evidence for normality but it does show a bell curve shape that can be indicative of a normal distribution.
### Figure 2 - Sample Data Normal Fit vs KDE
This figure compares the Kernel Density Estimate of the provided data to a normal curve fitted over the data. This graph like the last one helps show that bell curve like shape does exist in the data but the KDE can sometimes over smooth some data that might impact the data's normality. With that concern in mind we look at the next plot.
### Figure 3 - Normal Q-Q Plot
This plot compares the quantiles of our sample data vs that of a normal distribution. The closer the blue circles (sample data) are to the red line (perfect normality) makes our assuption of normality stronger. In this case, other than some tail outliers, the circles match up very well to the line indicating a strong correlation between out data and a normal distribution.

### Quantitative Tests for Normality
Below we run 2 tests that look to see how likely it is that our data comes from a normal distribution. The null hypothesis is that there is no difference between our sample data and the normal distribution.

** Shapiro-Wilks Test for Normality **

* Test Statistic: **0.987**
* P-value: **0.23**
* Result: **We do not reject the null hypothesis**. 0.23 > 0.05. This test finds evidence of normality at the a=0.05 significance level.

```python
stats.shapiro(df.temperature)
```

** Anderson Test for Normality **

* Test Statistic: **0.52**
* Critical Value for Rejecting H0: **> 0.765**
* Result: **We do not reject the null hypothesis**. 0.52 < 0.765. This test finds evidence of normality at the a=0.05 significance level.

```python
stats.anderson(df.temperature, dist='norm')
```

## 3. Is the true population mean really 98.6 degrees F?

The short answer, probably not. The true population is made up of billions of people that have ever changing body temperatures. The chance that the average of everyone's body temperature at a given point in time is actually 98.6 degrees F is practically 0.


What we can explore with our sample data howerver is how good of an estimate is an average body temperature of 98.6 degrees. We have already determined it is plausible that the data follows a normal distribution. Even though our sample size of 160 is greater than 30 and we believe the samples collected are independent of one another we should use the T-statistic since we don't know the population's standard deviation. We can still use the Z-statistic since with this many samples the results will be similar but we will start with the T-statistic to develop a 95% confidence interval around our sample mean and see if 98.6 falls within that confidence interval. 

There is no need for a two sample test since we are looking to see how a sample statistic performs against our estimated population statistic.

___
Khan academy has a t vs z statistic video/discussion that is useful here: https://www.khanacademy.org/math/statistics-probability/significance-tests-one-sample/tests-about-population-mean/v/z-statistics-vs-t-statistics
___

```python
sample_mean = df.temperature.mean()
sample_stdev = df.temperature.std()
sample_count = len(df.temperature)

print 'Sample Mean:', sample_mean
print 'Sample Standard Dev:', sample_stdev
print 'Sample Count:', sample_count
```

#### T-Statistic

Our two-tailed 95% Confidence interval have the following T-scores with 159 degrees of freedom

```python
lower_t = -2.8467
upper_t = 2.8467
```

To calculate our confidence interval we need to follow the following equation:

$$C.I. = \bar x \pm \frac {2.8467 * \sigma_x}{\sqrt n}$$

```python
cv = (sample_mean + ((lower_t * sample_stdev) / math.sqrt(sample_count)), sample_mean + ((upper_t * sample_stdev) / math.sqrt(sample_count)))
cv
```

So based on this sample's confidence interval of (98.066 to 98.432) we would reject the population mean having a chance of being 98.6 degreees F.

#### Z-Statistic

Our two-tailed 95% Confidence interval have the following Z-scores 

```python
lower_z = -1.96
upper_z = 1.96
```

To calculate our confidence interval we need to follow the following equation:

$$C.I. = \bar x \pm \frac {1.96 * \sigma_x}{\sqrt n}$$

```python
cv = (sample_mean + ((lower_z * sample_stdev) / math.sqrt(sample_count)), sample_mean + ((upper_z * sample_stdev) / math.sqrt(sample_count)))
cv
```

So based on this sample's confidence interval of (98.123 to 98.375) we would also reject the population mean having a chance of being 98.6 degreees F.

## 4. At what temperature should we consider someone's temperature to be "abnormal"?

Based on the results from question 3, at a 95% confidence interval using the T-statistic we would consider any temperature above 98.43 or below 98.07 to be abnormal.

To determine if this difference is significant we will run a permutation test.
H0: There is no difference in average body temperature between males and females
H1: There is a difference in average body temperature between males and females

To test this null hypothesis we can shuffle together all the samples and create two new random groups of equal size to the original groups and calculate the difference in their mean values (our test statistic).

```python
class HypothesisTest(object):
    """Represents a hypothesis test."""

    def __init__(self, data):
        """Initializes.

        data: data in whatever form is relevant
        """
        self.data = data
        self.MakeModel()
        self.actual = self.TestStatistic(data)
        self.test_stats = None

    def PValue(self, iters=1000):
        """Computes the distribution of the test statistic and p-value.

        iters: number of iterations

        returns: float p-value
        """
        iters = float(iters)
        self.test_stats = np.array([self.TestStatistic(self.RunModel()) 
                                       for _ in range(int(iters))])

        count = float(sum(self.test_stats >= self.actual))
        return count / iters

    def MaxTestStat(self):
        """Returns the largest test statistic seen during simulations.
        """
        return max(self.test_stats)

    def PlotHist(self, label=None):
        """Draws a Cdf with vertical lines at the observed test stat.
        """
        ys, xs, patches = plt.hist(ht.test_stats, color='blue')
        plt.vlines(self.actual, 0, max(ys), linewidth=3, color='0.8')
        plt.xlabel('test statistic')
        plt.ylabel('count')

    def TestStatistic(self, data):
        """Computes the test statistic.

        data: data in whatever form is relevant        
        """
        raise UnimplementedMethodException()

    def MakeModel(self):
        """Build a model of the null hypothesis.
        """
        pass

    def RunModel(self):
        """Run the model of the null hypothesis.

        returns: simulated data
        """
        raise UnimplementedMethodException()
```

```python
class DiffMeansPermute(HypothesisTest):
    """Tests a difference in means by permutation."""

    def TestStatistic(self, data):
        """Computes the test statistic.

        data: data in whatever form is relevant        
        """
        group1, group2 = data
        test_stat = abs(group1.mean() - group2.mean())
        return test_stat

    def MakeModel(self):
        """Build a model of the null hypothesis.
        """
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))

    def RunModel(self):
        """Run the model of the null hypothesis.

        returns: simulated data
        """
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
```

Running this model many times we can see how the generated mean differences compare to the actual difference in male vs female average body temperatures. We can also compute a p-value by seeing how many times out of all the iterations the test statistic exceeded our actual sample mean difference.

```python
data = (male.temperature, female.temperature)
ht = DiffMeansPermute(data)
p_value = ht.PValue(iters=1000000)
print 'p-value =', p_value
print 'actual =', ht.actual
print 'ts max =', ht.MaxTestStat()
```

```python
ht.PlotHist()
```

Above we ran our model 1 million times and returned a p-value of 0.025. This means that only 2.5% of the time was the difference in means between the randomly created groups greater than or equal to our calculated mean difference between male and female test subjects. 


Based on a 95% confidence level we can reject the null hypothesis. Since it is such a remote chance that this difference in means can be seen randomly we have evidence based on this sample that if we re-ran the experiment we would see a significant difference in the mean body temperatures between male and female test subjects.

This means that yes, there is a significant difference in body temperatures between males and females.
