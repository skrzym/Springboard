## Matplotlib object-oriented API

Much more control over each aspect of a plot (or plots) over the simple Matlab api. Especially useful for plotting multiple plots at the same time (subplots).

1. Using numpy we generate our x,y data to plot
2. Let's create the figure instance and store a reference to it in the `fig` variable.
3. We create an axis instances `axes` using the `add_axes` method in the `Figure` class that we have stored in the `fig` variable.
4. Now we can call the `plot` method on our `axes` to render the data to our figure. The third argument '`r`' set's the plot's line color/style.
5. Here we use `set_xlabel`, `set_ylabel`, and `set_title` to add descriptive text to the plot.

```python
# 1
x = np.linspace(0,5,10)
y = x ** 2

# 2
fig = plt.figure()

# 3
axes = fig.add_axes([0.0,0.0,1.0,1.0]) # Values range between 0 and 1 - left, bottom, width, height

#4
axes.plot(x,y,'r')

#5
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title')
```

This programatic control over the figure allows us much more freedom to create some unique looking plots... like below.

```python
fig = plt.figure()

axes1 = fig.add_axes([0.0,0.0,0.5,0.5])
axes2 = fig.add_axes([0.5,0.5,0.5,0.5])

axes1.plot(x,y,'c')
axes2.plot(y,x,'m')

axes1.set_title('Plot 1')
axes2.set_title('Plot 2')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
```

Instead of needing to explicitly place and size the axes in the figure, we can use a helper function `subplots` to neatly do this for us.

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')
```
